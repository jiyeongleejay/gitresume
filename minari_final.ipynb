```python
def rev_to_word(raw_review ):
    review_text = BeautifulSoup(raw_review, 'html.parser').get_text()
    letters_only = re.sub('[^a-zA-Z]', ' ', review_text
    words = letters_only.lower().split()
    from nltk.corpus import stopwords
    stops = set(stopwords.words('english'))
    meaningful_words = [w for w in words if not w in stops]    
    meaningful_words = [w for w in meaningful_words if not w=='permalink']
    meaningful_words = [w for w in meaningful_words if not w=='help']
    meaningful_words = [w for w in meaningful_words if not w=='review']
    meaningful_words = [w for w in meaningful_words if not w=='found']
    stemming_words = [stemmer.stem(w) for w in meaningful_words]
    return( ' '.join(stemming_words) )
```

```python
def get_title_keywords(text):    
    min_count = 1   # 단어의 최소 출현 빈도수 (그래프 생성 시)
    max_length = 10 # 단어의 최대 길이
    wordrank_extractor = KRWordRank(min_count=min_count, max_length=max_length)
    beta = 0.85    # PageRank의 decaying factor beta
    max_iter = 10
    keyword_list = []
    try:
        keywords, rank, graph = wordrank_extractor.extract([text], beta, max_iter)
        for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:30]:
                keyword_list.append(word)
#        keyword_list = []
#        for key, value in keywords.items():
#            if value >= 1:
#                keyword_list.append(key)
#        return keyword_list
    except ValueError:
        pass
    
    return keyword_list
```

```python
import re
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU, Embedding
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import load_model
!pip install numpy==1.19.5
import numpy as np
```

```python
vocab_size = 10000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words = vocab_size)
```

```python
max_len = 500
X_train = pad_sequences(X_train, maxlen=max_len)
X_test = pad_sequences(X_test, maxlen=max_len)
```

```python
np.__version__
```

```python
model = Sequential()

model.add(Embedding(vocab_size, 100))
model.add(GRU(128))
model.add(Dense(1, activation='sigmoid'))
```

```python
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=4)
mc = ModelCheckpoint('GRU_model.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train, y_train, epochs=15, callbacks=[es, mc], batch_size=60, validation_split=0.2)
```

```python
loaded_model = load_model('GRU_model.h5')
print("\n 테스트 정확도: %.4f" % (loaded_model.evaluate(X_test, y_test)[1]))
```
